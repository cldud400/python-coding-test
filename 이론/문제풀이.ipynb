## Two Sum
- 주어진 리스트의 원소들을 이용해서 덧셈을 했을 때, 타겟을 만들 수 있는 두 숫자 인덱스를 리턴하는 문제

### 내가 푼 방식

```python
nums = [3,2,4]
target = 6
list_ = []
for i in range(len(nums)):
  for j in range(i+1, len(nums)):
    if nums[i] + nums[j] == target:
        list_.append(i)
        list_.append(j)

print(list_)
```

### 브루트포스 방식으로 풀이

```python
nums = [3,2,4]
target = 6

for i in range(len(nums)):
  for j in range(i+1,len(nums)):
    if nums[i] + nums[j] == target:
        print ([i,j])
```

### targer -n이 존재하는지 확인

```python
for i, n in enumerate(nums):
          tmp = target - n
          if tmp in nums[i+1:]:
            return (i, nums[i+1:].index(tmp) + (i+1))
```

### dictionary를 이용한 풀이

```python
nums = [3,2,4]
target = 6
numS = {}

for i, n in enumerate(nums):
  tmp = target - n
  if tmp in numS:
    print(numS[tmp], i)
  numS[n] = i
```

### 정렬이 되어있다는 가정 하에 투포인터 방식으로 풀이

```python
numbers = [-12, -11, 1, 3, 9]
target = -10
left = 0
right = len(numbers) - 1
while not left == right:
  if numbers[left] + numbers[right] > target:
    right -= 1
  elif numbers[left] + numbers[right] < target:
    left += 1
  else:
    print ([left+1, right+1])
    break
```

## 1, 2, 3 더하기(9095)

```python
n = int(input())

def sumNumber(n):
    if n == 1:   return 1
    elif n == 2: return 2
    elif n == 3: return 4
    else:
        return sumNumber(n-1) + sumNumber(n-2) + sumNumber(n-3)

for i in range(n):
    num = int(input())
    print(sumNumber(num))
```

```python
class Node:
  def __init__(self, value):
    self.value = value
    self.left = None
    self.right = None

class Tree:
  def __init__(self, value):
    self.head = Node(value)

  def addNode(self, value):
    node = self.head  #최상위 노드
    newNode = Node(value)
    while True:
      # if value == null:
      #   value = None

      if value < node.value:    #값이 현재 노드보다 작으면 왼쪽에 추가
        if node.left != None:   #왼쪽에 자식 노드가 있으면, 자식 노드를 검사
          node = node.left
        else: # 왼쪽에 자식 노드가 없으면 그냥 추가하면 된다
          node.left = newNode
          break
      else:  #값이 현재 노드보다 크면 오른쪽에 추가
        if node.right != None:
          node = node.right
        else:
          node.right = newNode
          break

    def search(self, value):
      node = self.head
      while node:
        if node.value == value:
          return value
        elif value < node.value:
          node = node.left
        else:
          node = node.right
      return False



  
```
