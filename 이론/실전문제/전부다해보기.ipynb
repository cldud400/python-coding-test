## 차량번호판1(16968)
- 가능한 모든 번호판을 구하자

```python
def solve(str, x, ans):
  if x == len(str):
    return print(ans)

  if str[x] == 'd':
    if x > 0 and str[x-1] == 'd':
      solve(str, x+1, ans*9)
    else:
      solve(str, x+1, ans*10)
  
  if str[x] == 'c':
    if x > 0 and str[x-1] == 'c':
      solve(str, x+1, ans*25)
    else:
      solve(str, x+1, ans*26)

a = input()
solve(a,0,1)
```

```python
def solve(a, idx, last):
  if idx == len(a): return 1

  start = ord('a') if a[idx] == 'c' else ord('0')
  end = ord('z') if a[idx] == 'c' else ord('9')

  ans = 0
  for i in range(start, end+1):
    if i!= last:
      ans += solve(a, idx+1, i)
  return ans

print(solve('dcdd', 0, ' '))
```

## 세친구(17089)

## 캐슬 디펜스(17135)
- M개의 칸에 궁수 3명을 배치하는 방법
  - 궁수를 배치할 수 있는 방법은 M^3

- 배치할 수 있는 모든 칸에 궁수를 배치하고, 시뮬레이션
- 행이 N개, 적은 최대 N번 이동
- 궁수가 공격할 수 있는 적을 찾으려면 전부 순회해야 한다.
- 적이 이동하는것 처리
- 각 턴을 처리하는데 걸리는 시간(적을 찾는 시간(n\*m) + 적이 이동하는 시간(n\*m))
- 게임이 종료 되려면, n\*(n\*m)
- 모든 방법을 검사 (n\*m)^5

```python
from copy import deepcopy
n, m, d = map(int,input().split())
a = [list(map(int,input().split())) for _ in range(n)]

# 시뮬레이션
def solve(i1,i2,i3):
    b = deepcopy(a)
    ans = 0
    while True:
        cnt = 0
        # 적의 수를 확인
        for i in range(n):
            for j in range(m):
                cnt += b[i][j]

        if cnt == 0: break

        # 한번에 처리할 수 있는 적의 수는 최대 3
        deleted = [(-1,-1)] * 3
        archer = [i1,i2,i3]

        # 궁수의 순서대로 처리
        for k in range(3):
            ni, nj = n, archer[k]
            xi, xj, dist = -1, -1, -1
            for j in range(m):
                for i in range(n):
                    if b[i][j] != 0:
                        di = abs(ni-i)
                        dj = abs(nj-j)
                        dd = di + dj
                        # 적이 궁수의 사정거리 내에 있어야 한다
                        # 제일 가까운 적 처리
                        if dd <= d:
                            if dist == -1 or dist > dd:
                                xi = i
                                xj = j
                                dist = dd
            # 처리 된 적 
            deleted[k] = (xi,xj)

        # 처리된 적의 수 확인
        for i,j in deleted:
            if i == -1: continue
            if b[i][j] != 0: 
                ans += 1
                b[i][j] = 0
        
        # 처리가 끝나면 적이 이동한다.
        for i in range(n-1,-1,-1):
            for j in range(m):
                if i == 0: b[i][j] = 0
                else: b[i][j] = b[i-1][j]

    return ans





# 가능한 궁수 배치는 m^3 for 3개로 처리
ans = 0
for i1 in range(m):
    for i2 in range(i1+1,m):
        for i3 in range(i2+1,m):
            ret = solve(i1,i2,i3)
            if ans < ret: ans = ret

print(ans)


```

## 야구(17281)
- 선수는 항상 9명, 1번은 항상 4번타자
  - 나머지 타순을 정해야 한다
  - 가능한 타순의 개수는 8!
- 한 이닝에는 아웃을 당하는 타자가 적어도 한명
  - 한 이닝에 퇴대 타석의 수는 3번
  - 최대 27번 타석
  - n이닝을 수행
  - 시뮬레이션에 걸리는 시간 (n*27)

```python
from itertools import permutations

n = int(input())

a = [list(map(int,input().split())) for _ in range(n)]

ans = 0

def solve(arr):
    batter = 0
    score = 0
    for inning in a:
        out = 0
        base1,base2,base3 = 0,0,0
        while out < 3:
            if inning[arr[batter]] == 0:
                out += 1
            elif inning[arr[batter]] == 1:
                score += base3
                base1, base2, base3 = 1, base1, base2
            elif inning[arr[batter]] == 2:
                score += (base2+base3)
                base1, base2, base3 = 0, 1, base1
            elif inning[arr[batter]] == 3:
                score += (base1+base2+base3)
                base1, base2, base3 = 0, 0, 1
            elif inning[arr[batter]] == 4:
                score += (1+base1+base2+base3)
                base1, base2, base3 = 0, 0, 0

            
            batter = (batter+1) % 9
    
    return score

for perm in permutations(range(1,9),8):
    perms = list(perm[:3]) + [0] + list(perm[3:])
    ans = max(ans, solve(perms))

print(ans)
```

```python
n = int(input())
a = [ list(map(int, input().split())) for _ in range(n) ]
player = [0] * 9
check = [False] * 9

def simulation():
  # print(player)
  now = 0
  score = 0
  for inning in range(n):
    # base(안타를 친 경우)
    b1, b2, b3 = 0, 0, 0

    # out count
    out = 0

    while out < 3:
      result = a[inning][player[now]]
      if result == 0: out += 1
      elif result == 1: 
        # 3루에 주자가 있는 경우 점수를 낼 수 있습니다. 
        score += b3

        # 안타를 친 경우 주자는 한 베이스씩 진루 합니다. 
        b3, b2, b1 = b2, b1, 1

      elif result == 2:
        score += (b3 + b2)
        b3, b2, b1 = b1, 1, 0
      elif result == 3:
        score += (b3 + b2 + b1)
        b3, b2, b1 = 1, 0, 0
      elif result == 4:
        score += (b3 + b2 + b1 + 1)
        b3, b2, b1 = 0, 0, 0
      now += 1
      if now == 9: now = 0
  return score

def solve(idx):
  # 선수 배치가 끝나면, 직접 시뮬레이션 해봅니다.
  if idx == 9: return simulation()

  # 4번 타자는 1번 선수로 정해져 있습니다. 
  if idx == 3: player[idx] = 0; return solve(idx+1)

  ans = 0
  # 1번 선수를 제외한 나머지 8명의 선수의 순번을 정해야 합니다. 
  for i in range(1, 9):
    # 이전에 선택한 선수라면 다시 선택하지 않습니다. 
    if check[i]: continue
    # 그렇지 않다면, 선수를 선택해주고
    check[i] = True
    player[idx] = i
    score = solve(idx+1)
    if ans < score: ans = score
    check[i] = False
    player[idx] = 0
  return ans
print( solve(0) )
```

## 게리맨더링 - 2(17779)
- N x N인 격자모양의 도시를 5개의 선거구로 나누는 문제
- 한 선거구에 포함되어 있는 구역은 모두 연결되어 있어야 한다.(BFS, DFS)
- 구역을 나눴을 때, 인구가 가장 많은 선거구와 인구가 가장 적은 선거구의 차이의 최소값을 찾는 문제
- 기준점(x,y), 경계의 길이(d1, d2)를 정해야 한다. 
  - 기준점은 $n^2$을 넘지 않는다. 
  - 경계는 대각선 모양이기 때문에 d1, d2는 N을 넘지 않는다. 
  - x, y, d1, d2의 조합은 N * N * N * N

- 선거구를 나눈 후에 인구의 합은 전체를 순회해야 합니다. 

```python
from collections import deque

n = int(input())
a = [ list(map(int, input().split())) for _ in range(n) ]

di = [0, 0, 1, -1]
dj = [1, -1, 0, 0]
def bfs(d, i, j, g):
  n = len(d)
  d[i][j] = g
  q = deque()
  q.append((i, j))
  while q:
    i, j = q.pop()
    for k in range(4):
      ni, nj = i + di[k], j + dj[k]
      if 0 <= ni < n and 0 <= nj < n:
        if d[ni][nj] == 0: 
          d[ni][nj] = g
          q.append((ni, nj))

def solve(a, x, y, d1, d2):
  
  d = [ [0] * n for _ in range(n) ]

  for i in range(0, d1+1):
    d[x+i][y-i] = 5
    d[x+d2+i][y+d2-i] = 5
  for i in range(0, d2+1):
    d[x+i][y+i] = 5
    d[x+d1+i][y-d1+i] = 5

  for j in range(0, y-d1):
    d[x+d1][j] = 3
  
  for i in range(0, x):
    d[i][y] = 1

  for j in range(y+d2+1, n):
    d[x+d2][j] = 2

  for i in range(x+d1+d2+1, n):
    d[i][y-d1+d2] = 4

  bfs(d, 0, 0, 1)
  bfs(d, 0, n-1, 2)
  bfs(d, n-1, 0, 3)
  bfs(d, n-1, n-1, 4)

  cnt = [0] * 5
  for i in range(n):
    for j in range(n):
      if d[i][j] == 0: d[i][j] = 5
      cnt[d[i][j]-1] += a[i][j]

  cnt.sort()
  return cnt[-1] - cnt[0]

ans = -1
for x in range(n):
  for y in range(n):
    for d1 in range(1, n):
      for d2 in range(1, n):
        if 0 <= y-d1 and y+d2 < n:
          if x+d1+d2 < n:
            ret = solve(a, x, y, d1, d2)
            if ans == -1 or ans > ret: ans = ret

print(ans)
```

## 파이프 옮기기 - 1(17070)

```python
n = int(input())
a = [ list(map(int, input().split())) for _ in range(n) ]

def solve(i, j, dir):
  # i, j가 (n,n)에 도달하면, 한가지 방법이 존재
  if i == n-1 and j == n-1: return 1

  ans = 0
  if dir == 0: # 가로
    if j + 1 < n and a[i][j+1] == 0:
      ans += solve(i, j+1, 0)
    if i + 1 < n and j + 1 < n and a[i][j+1] == 0 and a[i+1][j] == 0 and a[i+1][j+1] == 0:
      ans += solve(i+1, j+1, 1)
  elif dir == 1: # 대각선
    if j + 1 < n and a[i][j+1] == 0:
      ans += solve(i, j+1, 0)
    if i + 1 < n and a[i+1][j] == 0:
      ans += solve(i+1, j, 2)
    if i + 1 < n and j + 1 < n and a[i][j+1] == 0 and a[i+1][j] == 0 and a[i+1][j+1] == 0:
      ans += solve(i+1, j+1, 1)
  elif dir == 2: # 세로
    if i + 1 < n and a[i+1][j] == 0:
      ans += solve(i+1, j, 2)
    if i + 1 < n and j + 1 < n and a[i][j+1] == 0 and a[i+1][j] == 0 and a[i+1][j+1] == 0:
      ans += solve(i+1, j+1, 1)

  return ans

print( solve( 0, 1, 0))
```

## 파이프 옮기기 - 2(17069)
- 파이프 옮기기 -1번을 응용하면 됩니다. 
- 같은 방법으로 풀 수는 없습니다. 
  - 최대 32까지 늘어났기 때문에, 모든 방법을 전부 확인할 수 없습니다. 
  - `동적계획법`을 이용해서 풀이를 진행
  - (i, j) -> (n, n)으로 가는 방법의 수를 d[i][j]에 저장하면서 진행
  - 파이프의 방향도 중요하기 때문에, 방향도 같이 기록
    - d[i][j][dir]

```python
n = int(input())
a = [ list(map(int, input().split())) for _ in range(n) ]
d = [ [ [-1]*3 for _ in range(n)] for _ in range(n) ]

def solve(i, j, dir):
  # i, j가 (n,n)에 도달하면, 한가지 방법이 존재
  if i == n-1 and j == n-1: return 1
  ans = d[i][j][dir]
  if ans != -1: return ans

  ans = 0
  if dir == 0: # 가로
    if j + 1 < n and a[i][j+1] == 0:
      ans += solve(i, j+1, 0)
    if i + 1 < n and j + 1 < n and a[i][j+1] == 0 and a[i+1][j] == 0 and a[i+1][j+1] == 0:
      ans += solve(i+1, j+1, 1)
  elif dir == 1: # 대각선
    if j + 1 < n and a[i][j+1] == 0:
      ans += solve(i, j+1, 0)
    if i + 1 < n and a[i+1][j] == 0:
      ans += solve(i+1, j, 2)
    if i + 1 < n and j + 1 < n and a[i][j+1] == 0 and a[i+1][j] == 0 and a[i+1][j+1] == 0:
      ans += solve(i+1, j+1, 1)
  elif dir == 2: # 세로
    if i + 1 < n and a[i+1][j] == 0:
      ans += solve(i+1, j, 2)
    if i + 1 < n and j + 1 < n and a[i][j+1] == 0 and a[i+1][j] == 0 and a[i+1][j+1] == 0:
      ans += solve(i+1, j+1, 1)

  d[i][j][dir] = ans
  return ans

print( solve( 0, 1, 0))
```

## 괄호 추가하기(16638)

```python
n = int(input())
a = list(input())
m = (n+1)//2 -1
ans = None

for s in range(1<<m):
    ok = True
    for i in range(m-1):
        if (s & (1<<i)) > 0 and (s & (1<<(i+1))) > 0:
            ok = False
    if not ok: continue

    b = a[:]
    #우선순위를 고려해서 계산을 하는것 보다 그냥 괄호 씌워서 직접 계산하면 된다
    for i in range(m):
        if (s & (1<<i)) > 0:
            k = 2 * i + 1
            b[k-1] = '(' + b[k-1]
            b[k+1] = b[k+1] + ')'
    exp = ''.join(b)
    result = eval(exp)  #문자열을 그냥 계산해줌(3+8*7-9*2 = 41)

    if ans is None or ans < result: ans = result

print(ans)
```

## 계란으로 계란치기(16987)

```python
n = int(input())

a = [0] * n
w = [0] * n

for i in range(n):
    a[i], w[i] = map(int,input().split())

def solve(idx):
    # 깨뜨릴 계란을 선택
    # n개의 계란이 선택이 됐으면, 다 깨보면서 확인
    if idx == n:
        cnt = 0
        for i in range(n):
            if a[i] <= 0: cnt += 1
        return cnt
    # 계란이 깨졌거나, 깨지지 않은 다른 계란이 없으면 넘어간다
    if a[idx] <= 0: return solve(idx+1)
    
    ans = 0
    flag = False
    for j in range(n):
        i = idx
        if i == j: continue

        # 깨지지 않은 계란
        if a[j] > 0:
            flag = True
            a[i] -= w[j]
            a[j] -= w[i]
            ret = solve(idx+1)
            if ans < ret: ans = ret
            # 확인 했으면 원복
            a[i] += w[j]
            a[j] += w[i]

    if not flag: return solve(idx+1)
    return ans

print(solve(0))
```

## Baaaaaaaaaduk2 - Easy (16988)
- 놓을 수 있는 곳에 전부 돌을 놓는다
  - 중첩루프를 이용하면, 비어있는 모든 곳의 위치를 확인할 수있다.
- 돌을 놓고나서 죽일 수 있는 상대 돌의 개수를 구한다
 - BFS로 확인

```python
from collections import deque
di = [0, 0, 1, -1]
dj = [1, -1, 0, 0]

n, m = map(int,input().split())
a = [list(map(int,input().split())) for _ in range(n)]

def bfs():
    ans = 0
    check = [[False] * m for _ in range(n)]
    for i in range(n):
        for j in range(m):
            if a[i][j] == 2 and check[i][j] == False:
                dead = True
                q = deque()
                q.append((i,j))
                check[i][j] = True
                cur = 0
                while q:
                    cur += 1
                    x, y = q.popleft()
                    for k in range(4):
                        nx, ny = x + di[k], y + dj[k]
                        if 0 <= nx < n and 0 <= ny < m:
                            if a[nx][ny] == 0: dead = False
                            elif a[nx][ny] == 2 and check[nx][ny] == False:
                                q.append((nx,ny))
                                check[nx][ny] = True
                if dead: ans += cur
    return ans
                            

ans = 0

# 모든 위치에 돌을 놓는다
for i1 in range(n):
    for j1 in range(m):
        if a[i1][j1] !=0: continue # 빈공간이 아니면 돌을 놓을 수 없다
        for i2 in range(n):
            for j2 in range(m):
                if a[i2][j2] != 0: continue
                if i1 == i2 and j1 == j2: continue

                a[i1][j1] = 1
                a[i2][j2] = 1
                result = bfs()
                if ans < result: ans = result
                a[i1][j1] = 0
                a[i2][j2] = 0

print(ans)
```

## 색종이 붙이기(17136)

```python
n = 10
a = [ list(map(int, input().split())) for _ in range(n) ]
avail = [ 0, 5, 5, 5, 5, 5, ]

# idx번째 칸에 색종이를 붙여봅니다. 
def solve(idx):
  # 전체칸에 색종이를 전부 붙였는데 
  # 붙이지 못한 칸이 있다면 실패
  if idx == n * n:
    for i in range(n):
      for j in range(n):
        if a[i][j] == 1: return -1
    return 0

  # 해당 칸이 0이면 색종이 붙이지 않으므로, 다음칸으로
  i = idx // n
  j = idx % n
  if a[i][j] == 0: return solve(idx + 1)

  # 그렇지 않으면, 붙여 봅니다. .
  ans = -1
  for k in range(5, 0, -1):
    # 색종이가 남아 있으면
    if avail[k] > 0:
      # 붙였을 때 칸을 벗어나면 안됩니다. 
      if i + k - 1 < n and j + k - 1 < n:
          flag = True
          for x in range(i, i+k):
            for y in range(j, j+k):
              # 색종이를 붙이는 공간보다 색종이가 더 큰면 안됩니다. 
              if a[x][y] == 0: flag = False; break
            if flag == False: break

          # 색종이를 붙일 수 있다면
          if flag:
            # 해당 공간을 0으로 바꿔줍니다. 
            for x in range(i, i+k):
              for y in range(j, j+k):
                a[x][y] = 0
            # 해당 색종이는 최대 5장 까지만 사용이 가능 
            avail[k] -= 1
            # 다음 칸으로 이동
            ret = solve(idx + 1)
            # 체크가 완료되면, 다른 경우를 확인하기 위해서는 원복
            avail[k] += 1

            if ret != -1:
              if ans == -1 or ans > ret + 1: ans = ret + 1

            # 확인 완료되면 다시 원복 시켜줍니다. 
            for x in range(i, i+k):
              for y in range(j, j+k):
                a[x][y] = 1
  return ans
print(solve(0))
```

## 미로 탈출하기(17090)
- 밖으로 나가지 않는 경우
  - LR과 같은 경우
  - d[i][j] = -1(방문x), 0(탈출불가), 1(탈출가능)

```python
n, m = map( int,  input().split() )
a = [ input() for _ in range(n) ]
d = [ [-1] * m for _ in range(n) ]

def solve(i, j):
  if i < 0 or j < 0 or i >=n or j >= m: return 1
  if d[i][j] != -1: return d[i][j]
  
  d[i][j] = 0

  if a[i][j] == 'R': d[i][j] = solve(i, j+1)
  elif a[i][j] == 'L': d[i][j] = solve(i, j-1)
  elif a[i][j] == 'U': d[i][j] = solve(i-1, j)
  else: d[i][j] = solve(i+1, j)

  return d[i][j]

for i in range(n):
  for j in range(m):
    if d[i][j] == -1: solve(i, j)

ans = 0
for i in range(n):
  for j in range(m):
    if d[i][j] == 1: ans += 1
print(ans)
```
