# DAG(Directed Acyctlic Graph)
- 싸이클이 없는 방향이 있는 그래프를 DAG라고 한다
- 위상정렬(Toplogical Sort)
- 그래프상에 간선 a -> b, a가 b보다 먼저라는 의미 일 때, 정점의 순서를 찾는 알고리즘

- 다음과 같은 그래프가 있다
![](https://img1.daumcdn.net/thumb/R720x0.q80/?scode=mtistory2&fname=http%3A%2F%2Fcfile2.uf.tistory.com%2Fimage%2F22540E485881AB7D0A58DC)

- 위상정렬은 DAG에서만 가능하고, 싸이클이 있으면 안된다.
- 그래프의 간선을 모두 뒤집어 놓고, DFS를 수행
- 정점이 스택에서 빠져나오는, 순서를 기록하면, 위상정렬과 순서가 같아진다.
- 스택에서 빠져나오는건, 더 이상 방문할 수 있는 정접이 없다는 것이고, 위상정렬이 끝난다

## 줄 세우기(2252)

BFS

```python
import sys
from collections import deque, defaultdict

input = sys.stdin.readline
n, m = map(int,input().split())

a = defaultdict(list)
in_degree = [0] * (n+1)


for _ in range(m):
    i, j = map(int,input().split())
    # i -> j
    a[i].append(j)
    # j번째에 들어오는 간선의 수
    in_degree[j] += 1

q = deque()

# in_degree가 0인 노드가 시작노드
for i in range(1, n+1):
    if in_degree[i] == 0:
        q.append(i)
# BFS
while q:
    x = q.popleft()
    print(x, end = ' ')
    for i in a[x]:
        in_degree[i] -= 1
        if in_degree[i] == 0:
            q.append(i)
print()
```

DFS

```python
import sys
from collections import deque, defaultdict
sys.setrecursionlimit(10000000)

input = sys.stdin.readline
n, m = map(int,input().split())

a = defaultdict(list)
check = [False] * (n+1)


for _ in range(m):
    i, j = map(int,input().split())
    # i -> j
    a[j].append(i)


q = deque()


# DFS
def solve(x):
    check[x] = True
    for i in a[x]:
        if check[i] == False:
            solve(i)
    print(x, end = ' ')

for i in range(1, n+1):
    if check[i] == False:
        solve(i)
print()
```

## 작업(2056)
- 위상정렬의 응용

```python
import sys
from collections import deque, defaultdict

input = sys.stdin.readline
n = int(input())

a = defaultdict(list)
work = [0] * (n+1)

for x in range(1,n+1):
    i, *j = map(int,input().split())
    work[x] = i
    if work[1] == 0:
        continue
    for y in j:
        if x in a:
            a[x].append(y)
        else:
            a[x] = [y]

for i in range(1,n+1):
    if i in a:
        ret = 0
        for j in a[i]:
            ret = max(ret,work[j])
        work[i] += ret

print(max(work))
```

```python
import sys
from collections import deque, defaultdict

input = sys.stdin.readline
n = int(input())
a = defaultdict(list)

work = [0] * (n+1)
in_degree = [0] * (n+1)
d = [0] * (n+1)

for i in range(1,n+1):
    tmp = list(map(int,input().split()))
    work[i] = tmp[0]
    for j in tmp[2:]:
        a[j].append(i)
        in_degree[i] += 1

q = deque()

for i in range(1,n+1):
    if in_degree[i] == 0:
        q.append(i)
        d[i] = work[i]


while q:
    x = q.popleft()
    for i in a[x]:
        in_degree[i] -= 1
        if d[i] < d[x] + work[i]:
            d[i] = d[x] + work[i]
        if in_degree[i] == 0:
            q.append(i)

print(max(d))
```

## 숨바꼭질 5(17071)

```python
from collections import deque

n, k = map(int,input().split())

d = [-1] * 500001
if n == k: print(0); exit()
q = deque()
q.append(n)
d[n] = 0
t = 1
while True:
    k += t
    if k > 500000: break
    nq = deque()
    while q:
        x = q.popleft()
        for i in [x+1,x-1,x*2]:
            if 0 <= i <= 500000:
                if d[i] != t:
                    nq.append(i)
                    d[i] = t
    if d[k] == t:
        print(t)
        exit()
    t += 1
    q = nq
print(-1)
    
```

## 배달(1175)

```python
from collections import deque

di = [0,0,1,-1]
dj = [1,-1,0,0]
n, m = map(int,input().split())

a = [list(input()) for _ in range(n)]
d = [[[[-1]*4 for _ in range(4)]for _ in range(m)] for _ in range(n)]

q = deque()

x1,y1,x2,y2 = -1,-1,-1,-1
ans = -1

for i in range(n):
    for j in range(m):
        if a[i][j] == 'S':
            for k in range(4):
                q.append((i,j,k,0))
                d[i][j][k][0] = 0
        elif a[i][j] == 'C':
            if x1 == -1: x1, y1 = i, j
            else: x2, y2 = i, j


while q:
    i, j, dir, s = q.popleft()
    
    if s == 3: ans = d[i][j][dir][s]; break

    for k in range(4):
        if dir == k: continue
        ni, nj = i + di[k], j + dj[k]
        if 0 <= ni < n and 0 <= nj < m:
            if a[ni][nj] != '#':
                ns = s
                if a[ni][nj] == 'C':
                    # '|=' OR 연산자, '&=' AND 연산자
                    if ni == x1 and nj == y1: ns |= 1
                    else: ns |= 2

                if d[ni][nj][k][ns] == -1:
                    d[ni][nj][k][ns] = d[i][j][dir][s] + 1
                    q.append((ni,nj,k,ns))

print(ans)
```

## 체스판 여행 -1(16959)
- 재방문 가능
  - 재방문을 하면 최단거리가 아니다
  - BFS로 풀 수 있다

- (i,j,num,piece)
  - 말의 위치 : i,j
  - 현재 몇번까지 방문을 했는지 : num
  - piece : 말의 종류

```python
from collections import deque

# 나이트
di1 = [-1, 1, 2, 2, 1, -1, -2, -2]
dj1 = [2, 2, 1, -1, -2, -2, -1, 1]
# 룩
di2 = [0,0,1,-1]
dj2 = [1,-1,0,0]
# 비숍
di3 = [-1, 1, 1, -1]
dj3 = [1, 1, -1, -1]



n = int(input())
a = [list(map(int,(input().split()))) for _ in range(n)]
d = [[[[-1] * 3 for _ in range(n*n)] for _ in range(n)] for _ in range(n)]


q = deque()
for i in range(n):
    for j in range(n):
        a[i][j] -= 1
        if a[i][j] == 0:
            for k in range(3):
                d[i][j][0][k] = 0
                q.append((i,j,0,k))


ans = -1
while q:
    i,j,num,piece = q.popleft()
    if num == n*n -1:
        if ans == -1 or ans > d[i][j][num][piece]:
            ans = d[i][j][num][piece]
    
    for k in range(3):
        if piece == k: continue
        if d[i][j][num][k] == -1:
            d[i][j][num][k] = d[i][j][num][piece] + 1
            q.append((i,j,num,k))
    
    # 나이트
    if piece == 0:
        for k in range(8):
            ni,nj = i + di1[k], j + dj1[k]
            if 0 <= ni < n and 0 <= nj < n:
                nextN = num
                if a[ni][nj] == num + 1: nextN = num + 1
                if d[ni][nj][nextN][piece] == -1:
                    d[ni][nj][nextN][piece] = d[i][j][num][piece] + 1
                    q.append((ni,nj,nextN,piece))
    # 룩
    elif piece == 1:
        for k in range(4):
            L = 1
            while True:
                ni,nj = i + di2[k] * L, j + dj2[k] * L
                if 0 <= ni < n and 0 <= nj < n:
                    nextN = num
                    if a[ni][nj] == num + 1: nextN = num + 1
                    if d[ni][nj][nextN][piece] == -1:
                        d[ni][nj][nextN][piece] = d[i][j][num][piece] + 1
                        q.append((ni,nj,nextN,piece))
                else: break
                L += 1
    # 비숍
    else:
        for k in range(4):
            L = 1
            while True:
                ni,nj = i + di3[k] * L, j + dj3[k] * L
                if 0 <= ni < n and 0 <= nj < n:
                    nextN = num
                    if a[ni][nj] == num + 1: nextN = num + 1
                    if d[ni][nj][nextN][piece] == -1:
                        d[ni][nj][nextN][piece] = d[i][j][num][piece] + 1
                        q.append((ni,nj,nextN,piece))
                else: break
                L += 1

print(ans)
```

## 숨바꼭질 -2(12851)
- 동생을 찾는 가장 빠른시간을 구하는 문제
  - 그러한 방법의 개수도 구해야한다

```python
from collections import deque

n, k = map(int,input().split())

d = [-1] * 200000
check = [False] * 200000
cnt = [0] * 200000
q = deque()
q.append(n)
d[n] = 0
check[n] = True
cnt[n] = 1

while q:
    now = q.popleft()
    for next in [now-1, now+1, now * 2]:
        if 0 <= next < 200000:
            if not check[next]:
                q.append(next)
                check[next] = True
                d[next] = d[now] + 1
                cnt[next] = cnt[now]
            elif d[next] == d[now] + 1:
                cnt[next] += cnt[now]

print(d[k])
print(cnt[k])
```

## 확장게임(16920)
- 플레이어 별로 BFS를 해준다

```python
from collections import deque

di = [0,0,1,-1]
dj = [1,-1,0,0]

n, m, p = map(int,input().split())
s = [0] + list(map(int,input().split()))
a = [[0] * m for _ in range(n)]


# .(빈곳)= 0, #(벽) = -1
for i in range(n):
    line = input()
    for j in range(m):
        if line[j] == '.': a[i][j] = 0
        elif line[j] == '#': a[i][j] = -1
        else: a[i][j] = int(line[j])

q = [ deque() for _ in range(p+1)]
nq = [ deque() for _ in range(p+1)]

for i in range(n):
    for j in range(m):
        if a[i][j] > 0:
            q[a[i][j]].append((i,j))

while True:
    flag = False
    for player in range(1, p+1):
        d = [[0] * m for _ in range(n)]
        while q[player]:
            flag = True
            i, j = q[player].popleft()
            if d[i][j] == s[player]: nq[player].append((i,j))
            if a[i][j] > 0 and a[i][j] != player: continue
            a[i][j] = player
            for k in range(4):
                ni, nj = i + di[k], j + dj[k]
                if 0 <= ni < n and 0 <= nj < m:
                    if a[ni][nj] == 0:
                        d[ni][nj] = d[i][j] + 1
                        if d[ni][nj] <= s[player]:
                            a[ni][nj] = player
                            q[player].append((ni,nj))

        q[player] = nq[player]
        nq[player] = deque()
    if not flag: break

ans = [0] * (p+1)
for i in range(n):
    for j in range(m):
        if a[i][j] > 0:
            ans[a[i][j]] += 1
print(*ans[1:])
```

## 점프게임(15558)
- 1초마다 칸이 사라지지 않으면, 그냥 BFS
  - i번째 칸을 방문한 초 >= i이면, 방문할 수 있다.

```python
from collections import deque
n, k = map(int,input().split())

a = [input() for _ in range(2)]
dir = [(0,1),(0,-1),(1,k)]
d = [[-1] * n for _ in range(2)]

q = deque()
q.append((0,0))
d[0][0] = 0
flag = False

while q:
    i, j = q.popleft()
    for di, dj in dir:
        ni, nj = (i + di)%2, j + dj
        if nj >= n: flag = True; break

        if nj < 0: continue
        if d[ni][nj] != -1: continue
        if a[ni][nj] == '0': continue
        if nj < d[i][j] + 1: continue
        d[ni][nj] = d[i][j] + 1
        q.append((ni,nj))
        
    if flag: break

if flag:
    print(1)
else:
    print(0)
```
