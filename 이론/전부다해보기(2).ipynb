## N-Queen(9663)
- 백트래킹으로 풀기
- 퀸은 가로, 세로, 대각선으로 움직일 수 있다.
- 각 행별로, 퀸을 어디에 놓을지 결정

```python
def check(row, col):
  # 같은 열의 다른 행에 퀸이 없어야 한다
  for i in range(n):
    if i == row: continue
    if a[i][col]: return False

  # 대각선에는 퀸이 없어야 한다.(왼쪽, 오른쪽)
  x = row - 1
  y = col - 1
  while x >= 0 and y >= 0:
    if a[x][y]: return False
    x -= 1
    y -= 1
  
  x = row - 1
  y = col + 1
  while x >= 0 and y < n:
    if a[x][y]: return False
    x -= 1
    y += 1
  
  return True

# 매 행 별로 체크
def solve(row):

  if row == n:
    global ans
    ans += 1
    return
  
  for col in range(n):
    a[row][col] = True
    if check(row,col):
      solve(row+1)
    a[row][col] = False

    
n = 8
ans = 0
a = [[False]*n for _ in range(n)]


solve(0)
print(ans)
```

시간복잡도를 줄이려면 중복 계산을 피해야 한다
- 이미 체스판 위에 놓여진 퀸들이 있다면 , 매번 퀸의 위치를 구하지 않고, 미리 기록해 뒀다가 사용하면 중복 계산을 피할 수 있다
- 열과, 대각선(/,\\)에 대해서 미리 기록

```python
def check(row, col):
  if check_col[col]: return False
  if check_dig1[row+col]: return False
  if check_dig2[row-col + n-1]: return False
  return True

# 매 행 별로 체크
def solve(row):

  if row == n:
    global ans
    ans += 1
    return
  
  for col in range(n):
    if check(row, col):
      check_col[col] = True
      check_dig1[row+col] = True
      check_dig2[row-col + (n-1)] = True

      a[row][col] = True
      solve(row+1)

      check_col[col] = False
      check_dig1[row+col] = False
      check_dig2[row-col + (n-1)] = False
      a[row][col] = False

    
n = 8
ans = 0
a = [[False] * n for _ in range(n)]
check_col = [False] * n
check_dig1 = [False] * (2*n-1)
check_dig2 = [False] * (2*n-1)

solve(0)
print(ans)
```

## 부등호(2529)
- 만약에 부등호가 (< >)이라면
   - 가장 큰 수는 9,8,7로 구성이 되면 된다.
   - 가장 작은 수는 0,1,2로 구성이 되면 된다.

- 다음순열, 이전순열을 구하는 방법을 이용

```python
def next_permutations(a):

  size = len(a) - 1
  i = size
  # while i <= size and a[i-1] < a[i]:
  #   i += 1
  # i -= 1
  while i > 0 and a[i-1] >= a[i]:
    i -= 1

  if i <= 0: return False
  
  j = size
  while a[j] <= a[i-1]:
    j -= 1

  a[i-1], a[j] = a[j], a[i-1]

  j = size
  while i < j:
    a[i], a[j] = a[j], a[i]
    i += 1
    j -= 1
  
  return True
```

```python
def prev_permutations(a):
  size = len(a) - 1
  i = size
  while i > 0 and a[i-1] <= a[i]: # 다음 순열과 반대
    i -= 1
  if i <= 0: return False
  j = size
  while a[j] >= a[i-1]:   # 다음 순열과 반대
    j -= 1
  a[i-1], a[j] = a[j], a[i-1]
  j = size
  while i < j:
    a[i], a[j] = a[j], a[i]
    i += 1
    j -= 1
  return True
```

```python
def check(perm, sign):
  for i in range(len(sign)):
    if sign[i] == '>' and perm[i] < perm[i+1]: return False
    if sign[i] == '<' and perm[i] > perm[i+1]: return False
    return True

k = 2
sign = ['<', '>']
small = [ i for i in range(k+1)]
big = [ 9-i for i in range(k+1)]


while True:
  if check(small, sign): break
  if not next_permutations(small): break

while True:
  if check(big, sign): break
  if not prev_permutations(big): break



print(big)

print(small)
check(small, sign)

```

## 단어수학(1339)
- 파이썬으로는 풀 수 없다.(시간초과)

```python
n = 2
# a = [['G','C','F'],['A','C','D','E','B']]
a = [['A','A','A'],['A','A','A']]
alpha = [0] * 26



for i in range(len(a)):
  for j in range(len(a[i])):
    a[i][j] = ord(a[i][j]) - 65

for i in range(n):
  j = 0
  for w in a[i][::-1]:
    alpha[w] += (10 ** j)
    j += 1
    alpha.sort(reverse=True)
    ans, t = 0, 9
    for i in range(26):
      if alpha[i] == 0:
        break
      ans += (t * alpha[i])
      t -= 1


print(a)
```

```python
def next_permutations(a):

  size = len(a) - 1
  i = size
  # while i <= size and a[i-1] < a[i]:
  #   i += 1
  # i -= 1
  while i > 0 and a[i-1] >= a[i]:
    i -= 1

  if i <= 0: return False
  
  j = size
  while a[j] <= a[i-1]:
    j -= 1

  a[i-1], a[j] = a[j], a[i-1]

  j = size
  while i < j:
    a[i], a[j] = a[j], a[i]
    i += 1
    j -= 1
  
  return True

def solve(a, letters, d):
  m = len(letters)
  ans = 0
  alpha = dict()
  for i in range(m):
    alpha[letters[i]] = d[i]
  for s in a:
    cur = 0
    for x in s:
      cur = cur * 10 + alpha[x]
    ans += cur
  return ans
n = 2
a = ['']*n
ans = 0
letters = set()

a[0] = 'GCF'
a[1] = 'ACDEB'

letters = set(a[0])
letters |= set(a[1])
letters = list(letters)
print(letters)

m = len(letters)
d = [i for i in range(9,9-m,-1)]
d.sort()
print(solve(a, letters, d))

while True:
  ret = solve(a, letters, d)
  if ans < ret:
    ans = ret
  if not next_permutations(d):
    break

print(ans)
```
