```python
#함수만들기 1
import random
n = 10
sample = [random.randint(1, 10) for _ in range(n)] #랜덤한 1~10까지의 n개의 정수를 가진 리스트
def solve(list_):
  total = 0
  for i in list_:
    total += i
  return total

print(sample)
solve(sample)
```

```python
#함수만들기 2 (1자리 자연수, 2자리 자연수는 한수로 취급힌다)
# n = 100 이라면 한수는 99개

def isHansu(num):
  list_ = list(str(num)) #각 자리수를 리스트로 만들어서
  list_ = list(map(int, list_)) #리스트를 정수로 바꿔준다
  degree = []
  for i in range(len(list_) -1): #마지막 자리는 빼준다
    degree.append(list_[i] - list_[i+1]) #각 자리수의 차
  if min(degree) == max(degree): #리스트 안에서 최솟값 최댓값
    return True
  else:
    return False

n = 1000
cnt = 99
if len(str(n)) < 3:  # str로 바꿔서 자리수를 센다
  cnt = n
else:
  for i in range(100, n + 1):
    ret = isHansu(i)
    if ret:
      cnt += 1
print(cnt)
```

```python
# 1부터 n까지의 합을 구하는 방법
# 1. 상향식 접근(loop)
n = 1000
total = 0
for i in range(1, n+1):
  total += i
print(total)
```

```python
# 2. 상향식 접근(꼬리재귀(tail recursion))
# loop로 표현가능하면 꼬리재귀로도 가능해야 한다.

def tail_recursion(total, n):
  if n == 0:
    return total
  return tail_recursion(total + n, n - 1)
tail_recursion(0, 100)
```

```python
# 3. 재귀로 표현

def recursion(n):
  if n == 1:
    return n
  return n + recursion(n - 1)
recursion(100)
```

```python
# 꼬리재귀

def tail_recursion(n, m, total=1): #default 값을 넣어준다
  if m >= 1:
    return tail_recursion(n , m - 1, total * n)
  else:
    return total 
tail_recursion(2, 5)
```

```python
# 재귀
def recursion(n, m):
  if m >= 1:
    return n * recursion(n, m - 1)
  else:
    return 1
recursion(10, 3)
```

```python
#꼬리재귀함수

def russian(n, m, total=0):
  if n >= 1:
    if n % 2:
      return russian(n//2, m*2, total + m)
    else:
      return russian(n//2, m*2, total)
  else:
    return total
russian(123, 12)
```

```python
#재귀함수
def russian(n, m):
  if n >= 1:
    if n % 2:
      return m + russian(n//2, m*2)
    else:
      return russian(n//2, m*2)
  else:
    return 0
russian(123,12)
```

```python
#꼬리재귀
def fac(n, mul=1):
  if n:
    return fac(n-1, n*mul)
  else:
    return mul
fac(6)
```

```python
#재귀
def fac(n):
  if n:
    return n * fac(n-1)
  else:
    return 1 #곱셈의 항등원
fac(5)
```

```python
#피보나치 문제 꼬리재귀

def fibo(n, current = 0, prev = 1):
  if n > 0:
    return fibo(n-1, current + prev, current)
  else:
    return current
fibo(10)
```

```python
#피보나치 문제 재귀(트리재귀)

def fibo(n):
  if n<= 1: #0 or 1
    return n
  return fibo(n-1) + fibo(n-2)
fibo(10)
```

```python
#n = 3이라고 가정

n = 27
star = [[' ' for _ in range(n)] for _ in range(n)] #공백을 원소로 가지는 3x3리스트 생성

def fill(i, j, n):
  if n == 1:
    star[i][j] = '*' #0,0에 *찍는다 , 0,1에 *찍는다 ... 1,0에 *찍고... 
    return

  mul = n // 3  #n =  1로 만들어준다

  for a in range(3):  #최소단위가 3x3
    for b in range(3):
      if a == 1 and b == 1: #(1,1)에만 공백
        continue
      fill(i + mul * a, j + mul * b, mul)     #나머지는 * (재귀)

fill(0,0,n)

for line in star:    #리스트를 스트링으로 변환
  for j in line:
    print(j, end='')
  print()
```
