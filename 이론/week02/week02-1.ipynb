```python
def printf(string = '기본값'): #문법에서는 맞지 않지만 지원해준다 (함수를 실행해야 메모리에 저장되는데 실행되기전에 메모리에 저장이 되기때문)
  print(string)
printf()
```

```python
def printf(*string): #가변인자로 구분을 했을때는 출력이 가능
  print(string)

printf('hello', 'python') #가변인자로 구분했을 때는 여러개의 입력을 하나의 튜플로 출력
```

```python
def printf(*args, **kargs): #일반 가변인자, 키 값을 함께 전달해주는 가변인자
    print(args, kargs)

printf(1 ,2 , var1 = 'hello', var2 = 'python') #DICT로 출력
```

```python
def function(list_):
  list_.reverse()
  return list_

sample = [10, 20, 30]
function(sample) #원본 sample값도 바뀐다, 값 전체를 전달하지 않고, 주소만 전달
```

```python
def recursive(n):
  if n <= 1:
    return 1
  return n + recursive(n - 1)

recursive(10)
```
