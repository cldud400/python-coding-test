```
#별찍기 2차원리스트 ver
n = 5
list_list = [['' for _ in range(n)] for _ in range(n)] #빈칸을 원소로하는 5x5리스트

for i in range(n):
  for j in range(i+1):
    list_list[i][j] = '*'


for list_ in list_list:  #리스트를 스트링으로 변환
  for ele in list_:
    print(ele, end='')
  print()
```
