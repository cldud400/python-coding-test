## 시간 복잡도의 표현
- Big O Notation
    - 입력 데이터의 크기에 따른 시간이 얼마나 걸리는지 표현

```python
# 1부터 n까지의 총합을 구하는 코드
n = 10
total = 0
for i in range(1, n + 1):
  total += i
print(total)
```

- 시간 복잡도를 표현하게 되면 정확히 입력된 n의 값만큼 동작
- O(n) = O(10)



```python
# 1부터 n까지의 총합을 구하는 코드 - 2
n = 10
total = n * (n + 1) // 2
print(total)
```

- 시간 복잡도는 O(1)이 됩니다.
- 상수번 내에 끝나기 때문

## 가장 많이 사용되는 시간 복잡도

# 스택(stack)
- '쌓는다'는 뜻
- 한쪽이 막혀있고, 한쪽으로만 입/출력이 발생하기 때문에, 마치 그 모습이 데이터가 쌓이는 모습과 유사하다고 해서, stack
- LIFO(Last In First Out)
- push, pop, top/sp(최상위)




```python
import sys
stack = []
n = int(sys.stdin.readline())

for _ in range(n):
  cmd, _, value = sys.stdin.readline().partition(' ')
  cmd, value = cmd.strip(), value.strip() #srtip : 앞, 뒤 공백제거
  if cmd == 'push':
    stack.append(value)
  elif cmd == 'top':
    if len(stack): print(stack[-1])
    else: print(-1)
  elif cmd == 'size':
    print(len(stack))
  elif cmd == 'pop':
   if len(stack): print(stack.pop(-1))
   else: print(-1)
  elif cmd == 'empty':
    if len(stack): print(0)
    else: print(1)
```

```python
import sys
stack = []
t = int(sys.stdin.readline())

for _ in range(t):
  braket = sys.stdin.readline().strip()
  for x in braket:
    if x =='(':
      stack.append(1)
    else:
      if len(stack) == 0:
        print('NO')
        break
      else:
        stack.pop(-1)
if len(stack):
  print('NO')
else:
  print('YES')
```
