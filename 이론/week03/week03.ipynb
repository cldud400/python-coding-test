```python
class Person:

  #클래스 변수 = 클래스 내부, 메서드 외부에 정의된 변수
  _nation = 'korea'

  #객체를 생성하지 않고도 사용할 수 있는 메서드가 필요
  @staticmethod
  def staticMethod(): #self 인자로 갖지 않는다
    return Person._nation

  @classmethod
  def classMethod(cls): #클래스를 인자로 하는 cls를 기본 파라미터로 갖는다
    return cls._nation

  def __init__(self, age = 0, name = ''):
    self._age = age       #class 내부 변수를 비공개 속성으로 정의하고 getter를 제공하는 것이 일반적
    self._name = name

  def method(self):
    print('this is method')
  
  def info(self):
    print('age: {}, name: {}'.format(self._age, self._name))

#클래스 내부에서 비공개 속성에 접근하는 방법
  @property           #getter을 뜻하는 decorate
  def age(self):      #속성을 가져오는 method, '_이름'에 대한 getter는'이름'으로 사용하는게 일반적
    return self._age
  @age.setter         #decorate, getter이 있어야 setter도 쓸 수 있다.
  def age(self, age): #속성을 update하는 method
    self._age = age

  @property
  def name(self):
    return self._name
  @name.setter
  def name(self, name):
    self._name = name
```

```python
장동건 = Person(20, '장동건')
장동건.info()
장동건.age       #공개된 속성에 접근하듯이 사용이 가능 , 장동건._age()와 같은 메서드 호출이다
장동건.age = 30  #장동건.age(30), 메서드 호출이다
장동건.name = '원빈'
장동건.info()
```

```python
class Student(Person): #상속
  
  _nation = 'English'

  #부모 클래스의 메서드 호출 super().

  #method overriding
  def __init__(self, age, name, school=''):
    super().__init__(age,name)
    self._school = school


  #method overriding
  def method(self):   #같은 이름의 메서드를 만들어서 바꿔준다
    #Person.method(self) #부모 클래스의 메서드를 호출하는 방법(self를 돌려줘야한다) 잘 사용하지 않는 방법
    super().method()    # 보통 사용하는 방법(self를 안 돌려둬도 됨)
    print('this is child\'s method')

  #method overriding
  def info(self):
    super().info()
    print('school: {}'.format(self._school))
```

```python
장동건 = Student(20, '장동건', '메가아이티')
장동건.info()
장동건.method()
Student.staticMethod()
Student.classMethod()
```

```python
class A:
  def __init__(self):
    print('A Class init')

class B(A):
  def __init__(self):
    print('B Class init')
    super().__init__()
    #A.__init__(self)
    self.attr = 'B'

class C(A):
  def __init__(self):
    print('C Class init')
    super().__init__()
    #A.__init__(self)
    self.attr = 'C'
```

```python
class D(C, B): #다중상속 우선순위 C -> B
  def __init__(self):
    print('D Class init')
    #B.__init__(self)
    #C.__init__(self)
    super().__init__()   # super을 사용하면  A를 두번 호출하지 않는다
```
