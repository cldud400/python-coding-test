### 거품정렬(Bubble Sort)

### 삽입정렬(insertion sort)
- 인메모리 방식은 조금 어려울 수 있으니, 정렬된 리스트를 따로 구하는 방식으로 한다.

```python
import random
unsort = [x for x in range(1, 10)]
random.shuffle(unsort)
display(unsort)
```

#### 가장 쉬운 삽입정렬

```python
# 가장 쉬운 삽입정렬
sort = []
print(sort, unsort)

for pick in unsort:
  print(pick)
  i = 0
  for i in range(len(sort)):
    if sort[i] > pick:
      sort.insert(i, pick)
      break
  if i >= len(sort) - 1:      # 마지막 원소는 안 들어갈 수 있어서 마지막에 넣어준다
    sort.append(pick)
  print(sort, unsort)
  
```

#### while 루프와 슬라이스를 이용한 구현

```python
# x를 주어진 list_에 적당한 위치에 삽입하는 역할
def insert(x, list_):
  # while 루프를 사용하려면 인덱싱이 까다롭다
  # 리스트를 슬라이싱 해서 항상 0번째 인덱스를 참조할 수 있게 한다.
  sort = []
  while list_ != []:
    if x <= list_[0]:  # 정렬할 필요가 없을 경우
      sort.append(x)
      return sort + list_
    else:
      sort.append(list_[0])
      list_ = list_[1:]   # 0번째 원소를 잘라준다   
  sort.append(x)
  return sort


def insertion_sort(list_):
  sort = []
  while list_ != []:
    list_, sort = list_[1:], insert(list_[0], sort)
  return sort

print(unsort)
insertion_sort(unsort)
```

#### 꼬리재귀를 이용한 방법
- insert(), insertion_sort()를 꼬리재귀로 바꾼다

```python
def insert(x, list_, sort):
  if list_ != []:    # 기저조건
    if x <= list_[0]:  
      sort.append(x)
      return sort + list_
      
    else:
      sort.append(list_[0])
      return insert(x, list_[1:], sort)
      
  else:
    sort.append(x)
    return sort


def insertion_sort(list_, sort):
  if list_ != []:    # 기저조건
    return insertion_sort(list_[1:], insert(list_[0], sort, []))
  else:
    return sort

# insert(5, [1,2,3,6], [])
print(unsort)
insertion_sort(unsort, [])
```

#### 재귀함수

```python
def insert(x, list_):
  if list_ != []:    # 기저조건
    if x <= list_[0]:  
      return [x] + list_
      
    else:
      return [list_[0]] + insert(x, list_[1:])
      
  else:
    return [x]


def insertion_sort(list_):
  if list_ != []:    # 기저조건
    return insert(list_[0], insertion_sort(list_[1:]))
  else:
    return []

print(unsort)
insertion_sort(unsort)
```

#### in-memory 방식
- for 반복문 이용 다른 리스트 없이 주어진 리스트 내에서 정렬이 되도록 구현

```python
print(unsort)
size = len(unsort)

for i in range(size):
  for j in range(i):
    if unsort[i] < unsort[j]:
      unsort.insert(j, unsort.pop(i))
      break
  print(unsort)
```

주어진 리스트에서 최솟값의 위치를 돌려주는 함수를 정의

```python
def minIndex(list_):
  min_ = list_[0]
  idx = 0
  for i in range(1, len(list_)):
    if list_[i] < min_:
     min_ = list_[i]
     idx = i
  return idx
```

#### in-memory 방식 (리스트 하나로 하기)

```python
def minIndex(list_, start, end):
  min_ = list_[start]
  idx = start
  for i in range(start, end):
    if list_[i] < min_:
     min_ = list_[i]
     idx = i
  return idx
```

```python
print(unsort)
size = len(unsort)
for i in range(size):
  idx = minIndex(unsort, i, size)
  unsort[i], unsort[idx] = unsort[idx], unsort[i]
print(unsort)
```

### 수 정렬

```python
import sys


n = int(sys.stdin.readline().rstrip())
sort = []
for _ in range(n):
    m = int(sys.stdin.readline().rstrip())
    if len(sort) == 0:
        sort.append(m)
        continue
    flag = True
    for i in range(len(sort)):
        if sort[i] > m:
            sort.insert(i, m)
            flag = False
            break
    if flag:            # 작은값이 없으면 맨뒤에 추가
        sort.append(m)

for x in sort:
    print(x)
```

### 합병정렬(Merge Sort)
- 재귀를 사용하는 대표적인 알고리즘
- 분할-정복 기법이라고 부른다.
  - 더 이상 분할할 수 없을 때까지, 분할하고
  - 정렬하면서 다시 합치는 방법을 사용

```python
import random
unsort = [x for x in range(1, 11)]
random.shuffle(unsort)
display(unsort)
```

합치는 함수

```python
# 리스트 슬라이스
def merge(left, right):
    sort = []
    while not (left == [] or right == []):
        if left[0] <= right[0]:
            sort.append(left[0])
            left = left[1:]
        else:
            sort.append(right[0])
            right = right[1:]
    return sort + left + right
```

```python
def merge(left, right):
  sort = []
  i, j = 0, 0
  left_size = len(left)
  right_size = len(right)
  while left_size > i and right_size > j:
    if left[i] > right[j]:
      sort.append(right[j])
      j += 1
    else:
      sort.append(left[i])
      i += 1

  if left_size > i:
    sort.extend(left[i:])
  elif right_size > j:
    sort.extend(right[j:])
    
  return sort
```

분할 함수

```python
def merge_sort(unsort):
  size = len(unsort)
  if size > 1:
    mid = size // 2
    return merge(merge_sort(unsort[:mid]), merge_sort(unsort[mid:]))
  else:
    return unsort
```

### 퀵 정렬
- 적당한 pivot이 잘 선택되어야 가장 잘 동작하는 알고리즘
  - 만약에 pivot이 가장 작은 값이나 가장 큰 값이 선택되면 너무 비효율적

```python
def split(pivot, unsort):
  left, right = [], []
  for x in unsort:
    if x <= pivot:
      left.append(x)
    else:
      right.append(x)
  return (left, right)
```

```python
def quick_sort(unsort):
  size = len(unsort)
  if size > 1:
    pivot = unsort[0]   #unsort의 가장 첫번째를 pivot로 사용
    left, right = split(pivot, unsort[1:])
    return quick_sort(left) + [pivot] + quick_sort(right)
  else:
    return unsort
```
