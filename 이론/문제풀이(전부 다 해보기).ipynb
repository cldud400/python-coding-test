# 전부 다 해보기
- brute force
- 문제에 대해서 전부 다 해볼 수 있는 경우의 수를 다 따져서 처리
- 경우의 수가 너무 많다면, 경우의 수를 줄일 수 있는 방법을 찾아서 적용


## 리모컨(1107)

```python
n = int(input())
m = int(input())
fail = input().split()
button = [x for x in range(10)]
for x in fail:
  button.remove(int(x))

def checkNumber(n):
  strs = str(n)
  cnt = 0
  for number in strs:
    if int(number) not in button:
      return False
    cnt += 1
  return cnt

cnt = abs(100 - n)        # abs는 절댓값을 의미한다
for num in range(1000001): # 0부터 10000000까지 돌려서
  ret = checkNumber(num)
  if ret:                 # num = 고장난 버튼을 제외한 버튼으로 만들수 있는 수
    if cnt > ret + abs(n-num): # n - num이 가장 작을 때를 찾는다
      cnt = ret + abs(n-num)

print(cnt)
```

## 카잉달력(6064)
- 날짜계산(1476)과 유사한 문제
- 모든 경우를 전부 계산하면 풀 수 있다.
  - 그러나 최대 4만 * 4만 = 16억번 돌려야 한다.
  - 적당한 패턴을 찾아야한다
- 즉, 이런경우 M만큼 건너 뛰면서, y찾기만 하면 된다
  - 그러면 시간복잡도는 $O(MN)$에서 $O(N)$번 만큼 줄어든다.

```python
import sys

t = int(sys.stdin.readline().rstrip())

for _ in range(t):
    m, n, x, y = map(int, sys.stdin.readline().rstrip().split())
    # 나누어 떨어지는 경우 즉, x, y가 0이되는 경우가 존재
    # 범위를 0 <= x,y <= M-1, N-1
    x -= 1
    y -= 1

    k = x

    while k < m*n:
        if k % n == y:
            print(k+1)
            break
        
        k += m
    
    else:
        print(-1)
```

### 내가 푼 방법

```python
n = int(input())
m = n
cnt = 0
while m >= 1:
  m /= 10
  cnt += 1

def counter(i):
  if i > 0:
    return (10**i - 10**(i-1)) * (i) + counter(i-1)
  else:
    return 0


def calc(n, x):
  if x > 0:
    return (n - 10**(x-1) + 1) * x + counter(x-1)
  else:
    return 0

calc(n, cnt)
```

# 조합 , 순열
- 완전 탐색의 대표적인 예
- 서로다른 n개의 원소에서 r개의 원소를 뽑는 방법
- 조합(combination)
  - 서로 다른 n개의 원소에서 r개의 원소를 뽑는 방법
  - 순서에 상관없다
  - 중복의 허용 여부에 따라서 조합의 갯수가 달라진다.

- 순열(permutation)
  - 서로 다른 n개의 원소에서 r개의 원소를 뽑는 방법
  - 순서가 상관이 있는 경우
  - 중복의 허용 여부에 따라서 순열의 갯수가 달라진다.

## 중복조합
- 알파벳 소문자를 이용해서 만들 수 있는 r개의 원소를 뽑는 방법
  - 알파벳 소문자를 가지고 중복된 선택이 가능한 순서를 고려하지 않는 경우의 수

### r이 1인 경우

```python
alpha = list('abcdefghijklmnopqrstuvwxyz')
for i in alpha:
  for j in alpha:
    print('{}{}'.format(i,j), end = ' ')
  print()
```

```python
alpha = list('abcdefghijklmnopqrstuvwxyz')
for i in alpha:
  for j in alpha:
    for k in alpha:
      print('{}{}{}'.format(i,j,k), end = ' ')
    print()
  print()
```

- 조합으로 풀었던 문제
 재귀적으로 풀어보기
  - 1부터 k까지의 수열 중에서 해당 수를 선택할지 말지 두 가지의 경우로 볼 수 있다.
  - 6개의 수가 만들어지면 정답
  - 그렇지 않으면 불가능한 경우
  - 둘 다 아니면 다음경우 호출


```python
def solve(a, i, lotto):

  # 6개의 수가 만들어지는 경우: 정답
  if len(lotto) == 6:
    print(' '.join(map(str, lotto)))
    return

  # 6개의 수가 만들어지지 않는 경우 : 불가능
  if i == len(a):
    return

  # i번째 수를 선택하는 경우
  solve(a, i+1, lotto+[a[i]])
  # i번째 수를 선탣하지 않는 경우
  solve(a, i+1, lotto)

n, *a = 7,1,2,3,4,5,6,7
solve(a,0,[])
```

## 연산자 끼워넣기(14888)
- 연산자만 고려하자
  - '+'를 사용하는경우
  - '-'를 사용하는경우
  - '*'를 사용하는경우
  - '/'를 사용하는경우

각각 나눠서 재귀


```python
def solve(a, i, result, plus, minus, mul, div):
  global min, max
  if i == n:
    if min > result:
      min = result
    if max < result:
      max = result
    return

  if plus:
    solve(a, i+1, result+a[i], plus-1, minus, mul, div)
  if minus:
    solve(a, i+1, result-a[i], plus, minus-1, mul, div)
  if mul:
    solve(a, i+1, result*a[i], plus, minus, mul-1, div)
  if div:
    if result >= 0:
      solve(a, i+1, result//a[i], plus, minus, mul, div-1)
    else:
      solve(a, i+1, -(-result//a[i]), plus, minus, mul, div-1)
  

n = 3
a = [3,4,5]
plus, minus, mul, div = 1,0,1,0
min = 1000000000000
max = -1000000000000

solve(a, 1, a[0], plus, minus, mul, div)
print(max)
print(min)
```

## 두 동전(16197)
- 최대 10번 동안에 동전이 하나만 떨어지는 경우
  - 정답을 찾은 경우(동전이 하나만 떨어진 경우)
  - 불가능한 경우(10번이 넘었거나, 동전이 둘 다 떨어진 경우)
  - 다음경우(위치를 이동)

```python
n, m = 6,2
a = [['.','#'],['.','#'],['.','#'],['o','#'],['o','#'],['#','#']]

# n, m = map(int,(input().split()))
# a = []
# for _ in range(n):
#     a.append(list(input().rstrip()))

x1,y1,x2,y2 = -1,-1,-1,-1
dx = [0,0,1,-1]
dy = [1,-1,0,0]

for i in range(n):
  for j in range(m):
    if a[i][j] == 'o':
      if x1 == -1:
        x1,y1 = i,j
      else:
        x2,y2 = i,j
        a[i][j] = '.'

def solve(s,x1,y1,x2,y2):
  if s > 10:
    return -1
  
  fall1 = False
  fall2 = False
  if x1 < 0 or x1 >= n or y1 < 0 or y1 >= m:
    fall1 = True
  if x2 < 0 or x2 >= n or y2 < 0 or y2 >= m:
    fall2 = True

  if fall1 and fall2:
    return -1
  
  if fall1 or fall2:
    return s
  
  ans = -1
  for k in range(4):
    nx1, ny1 = x1 + dx[k], y1 + dy[k]
    nx2, ny2 = x2 + dx[k], y2 + dy[k]
    if 0 <= nx1 < n and 0 <= ny1 < m and a[nx1][ny1] == '#':
      nx1, ny1 = x1, y1
    if 0 <= nx2 < n and 0 <= ny2 < m and a[nx2][ny2] == '#':
      nx2, ny2 = x2, y2
    ret = solve(s+1,nx1,ny1,nx2,ny2)
    if ret == -1:
      continue
    if ans == -1 or ans > ret:
      ans = ret
  return ans

print(solve(0,x1,y1,x2,y2))
```

## 에너지 모으기(16198)

```python
def solve(a):

  if len(a) == 2:
    return 0
  
  ans =0

  for i in range(1,len(a)-1):
    energy = a[i-1] * a[i+1]
    b = a[:i] + a[i+1:]
    energy += solve(b)
    if ans < energy: ans = energy
  return ans



n = 4
a = [1,2,3,4]
print(solve(a))

```
