# 조합 , 순열
- 완전 탐색의 대표적인 예
- 서로다른 n개의 원소에서 r개의 원소를 뽑는 방법
- 조합(combination)
  - 서로 다른 n개의 원소에서 r개의 원소를 뽑는 방법
  - 순서에 상관없다
  - 중복의 허용 여부에 따라서 조합의 갯수가 달라진다.

  - 순서가 상관이 있는 경우
  - 중복의 허용 여부에 따라서 순열의 갯수가 달라진다.

# 조합과 순열(정리)
문제를 읽고, 조합을 적용해야 하는 문제인지, 순열을 적용해야 하는 문제인지 판단하는게 어렵다

1. 조합(combination)
  - n개의 원소에서 r개의 원소를 뽑는 방법
  - 순서와 상관없이 r개의 뽑는 방법
    - (1,2)와 (2,1)은 같은 조합
  - 중복된 값을 허용하는 경우
    - 중복된 값을 허용하는 경우를 복원추출
    - 중복된 값을 허용하지 않는 경우를 비복원추출


2. 순열(permutation)
  - n개의 원소에서 r개의 원소를 뽑는 방법
  - 순서가 중요하다
    - 1을 먼저 뽑는 경우와 2를 먼저 뽑는 경우는 다른 경우이다.
    - (1,2)와 (2,1)은 다른 조합
  - 중복된 값을 허용하는 경우도 있다.
    - 나올 수 있는 모든 경우의 수

```python
from itertools import combinations, combinations_with_replacement, permutations, product
```

```python
# 조합(비복원추출)
strs = ['a', 'b', 'c']
for x in combinations(strs,2):
  print(*x)
```

```python
# 순서 상관 없이 2개의 원소를 뽑는 비복원 추출
strs = ['a', 'b', 'c']
choice = [False, False, False] # 이미 선택된 원소를 다시 선택하지 않도록 하기 위해서
selected = ['', '']     # 만들고자 하는 원소의 개수

def combi(i, j, n, m):
  if i == m: 
    print(*selected)
    return

  for k in range(j, n):
    if choice[k]: continue
    choice[k] = True
    selected[i] = strs[k]
    combi(i + 1, k + 1, n, m)
    choice[k] = False 

combi(0,0,3,2)
```

## NM과 K -1 (18290)
- 전부 다 해보기
  - 시간복잡도 $O(NMC4)$
  - 즉, 100개의 경우 중에 4개의 조합을 고르는 문제 $100C4$
  - 3,921,225가지의 경우가 생긴다
  - (r-1, c), (r+1, c), (r, c-1), (r, c+1)

```python
n, m, k = map(int, input().split())
arr = [ list(map(int, input().split())) for _ in range(n)]
# 이전에 선택한 칸이 또 선택되지 않도록
choice = [ [False for _ in range(m)] for _ in range(n)]
ssum = -99999
```

```python
# tail recursion(꼬리재귀)
def solve(r, sum):
  global ssum
  if r == k:
    if ssum < sum: ssum = sum
    return

  for i in range(n):
    for j in range(m):
      # 비복원추출
      if choice[i][j]: continue

      # 인접하지 않은 칸이 계산 가능한 칸인지 확인
      flag = True
      if 0 <= i - 1 < n and 0 <= j < m:
        if choice[i-1][j]: flag = False
      if 0 <= i + 1 < n and 0 <= j < m:
        if choice[i+1][j]: flag = False
      if 0 <= i < n and 0 <= j - 1  < m:
        if choice[i][j-1]: flag = False
      if 0 <= i < n and 0 <= j + 1 < m:
        if choice[i][j+1]: flag = False
      
      if flag:
        choice[i][j] = True
        solve(r + 1, sum + arr[i][j])
        choice[i][j] = False

n, m, k = map(int, input().split())
arr = [ list(map(int, input().split())) for _ in range(n)]
# 이전에 선택한 칸이 또 선택되지 않도록
choice = [ [False for _ in range(m)] for _ in range(n)]
ssum = -99999

solve(0,0)
print(ssum)
```

## 다음순열(10972)
- 사전순으로 정렬했을 때, 다음에 순열을 찾는 것이다.
- 파이썬은 모든 순열을 찾은 다음에 해당 순열의 이전 순열과 다음 순열을 확인 할 수 있다.

- 다음 순열을 구하는 알고리즘
1. a[i-1] < a[i]를 만족하는 가장 큰 i를 찾는다.
2. a[i-1] < a[j]를 만족하는 가장 큰 j를 찾는다.
3. a[i-1]와 a[j]를 교환한다.
4. a[i]부터 순열을 전부 뒤집어 준다.

```python
from itertools import permutations
n = int(input())
numbers = [x for x in range(1,n+1)]
end = tuple(numbers[::-1])
list_ = tuple(map(int, input().split()))
cnt = 0

for perm in permutations(numbers, n):
  cnt += 1
  if perm == list_:
    if perm == end:
      print(-1)
      break
    x = cnt
  
  if cnt == x+1:
    print(*perm)
    x = -2
    break
    
```

```python
def next_permutations(a):

  size = len(a) - 1
  i = size
  # while i <= size and a[i-1] < a[i]:
  #   i += 1
  # i -= 1
  while i > 0 and a[i-1] >= a[i]:
    i -= 1

  if i <= 0: return False
  
  j = size
  while a[j] <= a[i-1]:
    j -= 1

  a[i-1], a[j] = a[j], a[i-1]

  j = size
  while i < j:
    a[i], a[j] = a[j], a[i]
    i += 1
    j -= 1
  
  return True

import sys
n = int(sys.stdin.readline())
a = list(map(int, sys.stdin.readline().rstrip().split()))
if next_permutations(a):
  print(' '.join(map(str, a)))
else:
  print(-1)
```

## 이전 순열(10973)
- 다음 순열과 반대로 찾으면 된다

```python
def prev_permutations(a):

  size = len(a) - 1
  i = size
  
  while i > 0 and a[i-1] <= a[i]: # 다음 순열과 반대
    i -= 1

  if i <= 0: return False
  
  j = size
  while a[j] >= a[i-1]:   # 다음 순열과 반대
    j -= 1

  a[i-1], a[j] = a[j], a[i-1]

  j = size
  while i < j:
    a[i], a[j] = a[j], a[i]
    i += 1
    j -= 1
  
  return True

import sys
n = int(sys.stdin.readline())
a = list(map(int, sys.stdin.readline().rstrip().split()))
# n = int(input())
# a = list(map(int, input().split()))
if prev_permutations(a):
  print(*a)
else:
  print(-1)
```

## 외판원 순회 -2(10971)
- 최단 거리 알고리즘으로 유명
- 시간복잡도는 최대 $O(N*10!)$ = $3,628,800$
  - 전부 다 계산 충분히 가능

```python
n = int(input())
arr = [list(map(int,input().split())) for _ in range(n)]
ans = 99999999
from itertools import permutations
for perm in permutations(range(n), n):
  flag = True
  cost = 0
  for i in range(n-1):
    if arr[perm[i]][perm[i+1]] == 0:
      flag = False
      break
    else:
      cost += arr[perm[i]][perm[i+1]]
  if flag and arr[perm[-1]][perm[0]] != 0:
    cost += arr[perm[-1]][perm[0]]
    ans = min(ans,cost)

print(ans)
```
