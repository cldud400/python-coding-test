## 팰린드롬인지 확인하기(10988)

- 의외로 슬라이스는 빠르다

````python
read = input()

if read == read[::-1]:
  print(1)
else: print(0)
````

````python
# 리스트 형식
read = list(input())
flag = True

while len(read) > 1: #리스트의 크기가 홀수이면 1이남고 짝수면 0이남기 때문
  if read.pop(0) != read.pop():
    flag = False
    break

if flag: print(1)
else: print(0)
````

````python
# 함수형식
def isPalindrome(list_):
  while len(read) > 1:
    if read.pop(0) != read.pop():
      return 0
  return 1

read = list(input())
print(isPalindrome(read))
````

덱을 이용한 팰린드롬

````python
from collections import deque

read = deque(input())
flag = True

while len(read) > 1: #리스트의 크기가 홀수이면 1이남고 짝수면 0이남기 때문
  if read.popleft() != read.pop():
    flag = False
    break

if flag: print(1)
else: print(0)
````

문자열을 뒤집었을 때, 같은 문자열이 나오면 팰린드롬

````python
# 리스트 내부를 직접 변경
read = list(input())

left, right = 0, len(read) - 1
while left < right:
  read[left], read[right] = read[right], read[left]
  left += 1
  right -= 1
print(read)
````

````python
# 다른 언어를 사용할 때 문자열을 직접 뒤집어야 하는경우
def reverseString(list_):
  left, right = 0, len(list_) - 1
  while left < right:
    # list_[left], list_[rignt] = list_[right], list_[left]
    tmp = list_[left]
    list_[left] = list_[right]
    list_[right] = tmp
    left += 1
    right -= 1
  return ''.join(list_)

read = input()
if reverseString(list(read)) == read:
  print(1)
else: print(0)
````

````python
def reverseString(list_):
  left, right = 0, len(list_)-1
  while left < right:
    list_[left], list_[right] = list_[right], list_[left]
    left += 1
    right -= 1
  return ''.join(list_)

read = input()
if reverseString(list(read)) == read:
  print(1)
else: print(0)
````

- 투 포인터 방식으로 포인터를 확장하면서 가장 긴 팰린드롬을 찾아야하는 문제

````python
s = ['a', 'asdsfaf', 'asfagadgah']

print(max(s))
print(max(s, key=len))
````

````python
read = input()
size = len(read)
result = ''
def findPallindrome(left, right):
  while left >= 0 and right < len(read) and read[left] == read[right]:
    left -= 1
    right += 1
  return read[left + 1:right]
if size < 2:
  print(1)
else:
  for i in range(size - 1):
    result = max(result, findPallindrome(i, i+1), findPallindrome(i, i+2))
  print(len(result))
````

# 애너그램
- 알파벳의 순서를 바꿔서 새로운 단어를 만드는 방법
- 두 단어가 같은 알파벳을 사용해서 만들어진 단어인지 확인
  - 그룹-애너그램
- 애너그램인지 확인하는 제일 쉬운방법은 정렬
  - 애너그램 관계인 단어들은 정렬하면 같은 문자열이 나온다

## 애너그램(6996)

````python
t = int(input())

for _ in range(t):
  left, right = input().split()
  if sorted(left) == sorted(right):     # 자료를 정렬해 준다 sort는 기본값을 변경 sorted는 변경한 새로운 값을 제공
    print(f'{left} & {right} are anagrams.')
  else:
    print(f'{left} & {right} are NOT anagrams.')
````

## group-anagrams

````python
strs = ["eat","tea","tan","ate","nat","bat"]
dict_ = {}
for i in strs:
  key = ''.join(sorted(i))
  if key not in dict_:
    dict_[key] = []
  dict_[key].append(i)
  result = []
  for key in dict_:
      result.append(dict_[key])
     
print(result)
````

````python
strs = ["eat","tea","tan","ate","nat","bat"]
anagrams = dict()

for word in strs:
  key = ''.join(sorted(word))
  if key in anagrams.keys():
    anagrams[key].append(word)
  else:
    anagrams[key] = [word]

print(list(anagrams.values()))
````

````python
from collections import defaultdict
strs = ["eat","tea","tan","ate","nat","bat"]
anagrams = defaultdict(list)

for word in strs:
  key = ''.join(sorted(word))
  anagrams[key].append(word)
list(anagrams.values())
````
