## 부분수열의 합(14225)
- 부분수열을 비트마스크를 이용해서 만들어서 풀어보겠습니다. 

```python
n = 3
a = [5, 1, 2]
c = [False] * (n*100000)

for i in range(1 << n):
  s = 0
  for j in range(n):
    if (i & (1 << j) ):
  #     print(j, end=' ')
       s += a[j] 
  # print()
  c[s] = True

i = 1
while True:
  if c[i] == False: break
  i += 1
print(i)
```

## 가르침(1062)
- 26개의 알파벳 중에서 k개를 골라야 합니다. 
  - k-5개를 골라야 합니다. ('anta', 'tica')는 무조건 배워야 해서
  - 26-5개의 알파벳 중에서 k-5개를 고르는 문제

```python
# 고른 알파벳을 이용해서 읽을 수 있는 단어의 갯수
def count(mask, words):
  cnt = 0
  for word in words:
    if (word & ((1<<26)-1-mask)) == 0:
      cnt += 1
  return cnt

# i번째 알파벳을 고를지 말지, ... 
# k개를 고른다.
def solve(i, k, mask, words):
  if k < 0: return 0
  if i == 26: return count(mask, words)

  ans = 0
  # i번째 알파벳을 고르는 경우
  t1 = solve(i+1, k-1, mask | (1<<i), words)
  if ans < t1: ans = t1

  if i not in [ord('a')-ord('a'), ord('n')-ord('a'), ord('t')-ord('a'), ord('c')-ord('a'), ord('i')-ord('a')]:
    # i번째 알파벳을 고르지 않는 경우
    t2 = solve(i+1, k, mask, words)
    if ans < t2: ans = t2
  return ans
```

```python
n, k = 3, 6
words = [0] * n

# ['a', 'b', 'c', 'd', ...]
# [2^1, 2^2, 2^3, 2^4, ... ]

# 'a', 'n', 't', 'c', 'i' 는 무조건 배워야 하고 
# k-5개의 알파벳 중에서 골라야 합니다. 
for i, x in enumerate(['antarctica', 'antahellotica', 'antacartica']):
  for ch in x:
    words[i] |= (1 << (ord(ch)-ord('a')))

for word in words:
  print('{}: {:026b}'.format(word, word))

print( solve(0, k, 0, words) )
```

## 구슬탈출-2(13460)
- 의미없는 방향
  - 같은 방향으로 두 번 이상 이동하는 건 의미가 없다. 
  - 한 방향으로 이동한 다음에, 반대방향으로 이동하는 건 의미가 없다. 
  - 1번: 4가지, 2번: 2가지, 3번: 2가지

```python
class Marble:
  def __init__(self, moved, hole, i, j):
    self.moved = moved
    self.hole = hole
    self.i = i
    self.j = j

def simulate(a, k, i, j):
  n = len(a)
  m = len(a[0])

  if a[i][j] == '.': return Marble(False,False,i,j)
  moved = False
  while True:
    ni,nj = i+ di[k], j + dj[k]
    if ni < 0 or ni >= n or nj < 0 or nj >= m:
      return Marble(moved, False, i, j)

    ch = a[ni][nj]
    if ch == '#': return Marble(moved, False, i, j) 
    elif ch in 'RB': return Marble(moved, False, i, j) 
    elif ch  == '.':
        a[i][j], a[ni][nj] = a[ni][nj], a[i][j]
        i, j = ni, nj
        moved = True
    elif ch == 'O':
        a[i][j] = '.'
        moved = True
        return Marble(moved, True, i, j) 



def check(a, dirs):
  n = len(a)
  m = len(a[0])
  hi, hj = 0,0
  ri, rj = 0,0
  bi, bj = 0,0


  # 구멍, 빨간구슬, 파란구슬의 위치를 확인
  for i in range(n):
    for j in range(m):
      if a[i][j] == 'O': hi,hj = i, j
      if a[i][j] == 'R': ri,rj = i, j
      if a[i][j] == 'B': bi,bj = i, j

  cnt = 0
  # 방향에 따라서, 구슬을 굴려보면서 몇번만에 들어가는지 시뮬레이션 한다.
  for k in dirs:
    cnt += 1
    hole1 = hole2 = False
    while True:
      p1 = simulate(a, k, ri, rj)
      ri, rj = p1.i, p1.j
      p2 = simulate(a, k, bi, bj)
      bi, bj = p2.i, p2.j

      if not p1.moved and not p2.moved: break
      if p1.hole: hole1 = True
      if p2.hole: hole2 = True

    if hole2: return -1
    if hole1: return cnt
  return -1
    


def valid(dirs):
  # 왼쪽 오른쪽 위 아래를 왔다갔다 하는건 의미가 없다.
  # 같은 방향으로 가는 것도 의미 없다.
  l = len(dirs)
  for i in range(l-1):
    if dirs[i] == 0 and dirs[i+1] == 1: return False
    if dirs[i] == 1 and dirs[i+1] == 0: return False
    if dirs[i] == 2 and dirs[i+1] == 3: return False
    if dirs[i] == 3 and dirs[i+1] == 2: return False
    if dirs[i] == dirs[i+1]: return False
  return True

n, m = map(int,input().split())
board = [input() for _ in range(n)]
ans = -1


di = [0,0,1,-1]
dj = [1,-1,0,0]

for k in range( 1<< (10*2)):
  # 경우의 수
  dirs = [0] * 10
  for i in range(10): dirs[i] = (k & 3); k >>= 2
  if not valid(dirs): continue

  a = [list(row) for row in board]
  ret = check(a, dirs)
  if ret == -1: continue
  if ans == -1 or ans > ret: ans = ret


print(ans)
```
