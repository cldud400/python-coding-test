## 퇴사(14501)
- 결국 이 문제는 비복원 추출(조합)
  - T를 전부 합한날이 N보다 클 수는 없다.
  - 조합을 했을 때 전체 T의 합이 N보다 작거나 같은 경우만 확인
  - P의 값이 가장 큰 경우 찾으면 되는 문제

- 결국 N개의 원소에서 N! 만큼 조합한 경우라고 볼 수 있습니다. 

아래와 같은 방식으로는 이 문제를 풀 수 없다.

```python
def solve(choice):
  ssum = 0
  for i in choice:
    ssum += t[i]
  if ssum > n:
    return False
  
  cost = 0
  for i in choice:
    cost += p[i]
  
  print(choice, cost)
  return cost

n = 7
t = [3, 5, 1, 1, 2, 4, 2]
p = [10, 20, 10, 20, 15, 40, 200]
choice = [x for x in range(n-2)]
max = 0

from itertools import combinations
for i in range(1, n):
  for comb in combinations(choice, i):
    ret = solve(comb)
    if ret:
      if max < ret: max = ret
print(max)
```

백트래킹 방식으로 풀어야 합니다. 
- 재귀를 할 때, 더 이상 반복이 의미가 없는 경우가 생길 수 있습니다. 
- 그런 경우는 더 이상 반복을 진행하지 않고, 전부 다 해보는 방식을 백트래킹 이라고 합니다. 

```python
def solve(day, ssum):
  global cost
  if day > n: return 
  if day == n:
    cost = max(cost, ssum)
    return 

  # 1. 날짜를 선택하지 않는 경우
  solve(day+1, ssum)

  # 1. 날짜를 선택하는 경우 
  solve(day + t[day], ssum + p[day])

import sys
n = int(sys.stdin.readline())
t = [0 for _ in range(n)]
p = [0 for _ in range(n)]
for i in range(n):
  t[i], p[i] = map(int, sys.stdin.readline().split())

cost = 0
solve(0, 0)
print(cost)
```

## 맞춰봐(1248)
- 최대 10자리의 숫자가 입력
  - $O(21^{10})$ = 16,679,880,978,201
  - 경우의 수를 줄여아 한다.

```python
def check():
  for i in range(n):
    ssum = 0
    for j in range(i,n):
      ssum += ans[j]
      if signs[i][j] == '0':
        if ssum != 0: return False
      
      elif signs[i][j] == '+':
        if ssum <= 0: return False

      elif signs[i][j] == '-':
        if ssum >= 0: return False
  return True



def solve(i=0):
  if i == n:
    return check()
  
  for k in range(-10,11):
    ans[i] = k
    if solve(i+1): return True
  return False



n = 4
read = '-+0++++--+'
signs = [[0 for _ in range(n)] for _ in range(n)]
ans = [0] * n

k = 0
for i in range(n):
  for j in range(i,n):
    signs[i][j] = read[k]
    k += 1

solve()
print(ans)
```

경우의 수를 줄여아 한다.
- sign[i][i]에는 i번째의 부호가 들어있음
- 부호가 '+'이면 1 ~ 10까지만 해보면 된다.
- 부호가 '-'이면 -10 ~ -1까지만 하면 된다
- 0이면 0인 경우 밖에 없다.
- 시간복잡도는 $(10^{10})$ = 10,000,000,000

```python
def check():
  for i in range(n):
    ssum = 0
    for j in range(i,n):
      ssum += ans[j]
      if signs[i][j] == '0':
        if ssum != 0: return False
      
      elif signs[i][j] == '+':
        if ssum <= 0: return False

      elif signs[i][j] == '-':
        if ssum >= 0: return False
  return True



def solve(i=0):
  if i == n:
    return check()

  # 0인 경우 해당 자릿수는 0
  if signs[i][i] == '0':
    ans[i] = 0
    return solve(i+1)
  
  for k in range(1,11):
    if signs[i][i] == '-': k = -k
    ans[i] = k
    if solve(i+1): return True
  return False



n = 4
read = '-+0++++--+'
signs = [[0 for _ in range(n)] for _ in range(n)]
ans = [0] * n

k = 0
for i in range(n):
  for j in range(i,n):
    signs[i][j] = read[k]
    k += 1

solve()
print(ans)
```

4자리의 숫자가 전부 만들어지고 나서 체크를 하지 않고,
매번  체크를 해서 더 이상 의미없는 반복을 줄일 수 있다.

```python
def check(index):
  ssum = 0
  # 이미 앞에서 선택된 숫자는 제외하고 나머지 숫자들에 대해서 부호가 맞는지 확인
  for i in range(index, -1, -1):
    ssum += ans[i]
    if signs[i][index] == '0':
      if ssum != 0: return False
    
    elif signs[i][index] == '+':
      if ssum <= 0: return False

    elif signs[i][index] == '-':
      if ssum >= 0: return False
  return True



def solve(i=0):
  # 4자릿수가 전부 만들어 진 경우, 정답
  if i == n:
    return True

  # 0인 경우 해당 자릿수는 0
  if signs[i][i] == '0':
    ans[i] = 0
    return check(i) and solve(i+1)
  
  for k in range(1,11):
    if signs[i][i] == '-': k = -k
    ans[i] = k
    if check(i) and solve(i+1): return True
  return False



# n = 4
# read = '-+0++++--+'
import sys
n = int(sys.stdin.readline())
read = sys.stdin.readline().rstrip()
signs = [[0 for _ in range(n)] for _ in range(n)]
ans = [0] * n

k = 0
for i in range(n):
  for j in range(i,n):
    signs[i][j] = read[k]
    k += 1

solve()
print(*ans)
```
